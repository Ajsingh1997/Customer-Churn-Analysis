```python
ax= sns.countplot(x='gender', data = df, hue= 'Churn',palette={'Yes': 'red', 'No': 'green'})
sns.set(rc={'figure.figsize':(18,8)})
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title('Churn By Gender')
```

# Above count plot shows Churn based on Gender 

```python
#Calculate the percentage of churn for each gender category
churn_counts = df.groupby('gender')['Churn'].value_counts(normalize=True).unstack() * 100

# Create a stacked bar chart
ax = churn_counts.plot(kind='bar', stacked=True, figsize=(12, 6), 
                       color={'Yes': 'red', 'No': 'green'})

# Add percentage labels
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.1f%%', label_type='center', color='white', fontsize=12)

# Customize labels and title
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Percentage", fontsize=14)
plt.xticks(rotation=0)
plt.title("Churn Percentage by Gender", fontsize=16)
plt.legend(title="Churn", labels=["No (Stayed)", "Yes (Churned)"], loc="upper right")

# Show plot
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
           'PaperlessBilling', 'PaymentMethod']

# Set up the subplot grid
rows, cols = 4, 3  # Adjust based on the number of plots (4x3 grid for 12 plots)
fig, axes = plt.subplots(rows, cols, figsize=(18, 16))  # Define the figure size

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through the columns and create a countplot for each
for i, col in enumerate(columns):
    sns.countplot(x=df[col], ax=axes[i], palette={'Yes':'yellow', 'No':'Blue'}, hue=df['Churn'])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

#The visualizations show that customers without additional services like Online Security, Tech Support, and Backup tend to churn more. Fiber optic users, month-to-month contracts, and paperless billing customers have the highest churn rates. To improve retention, strategies should focus on long-term contracts, bundled services, and enhanced customer support.

#Customers with month-to-month contracts, electronic check payments, and paperless billing have higher churn rates. Those with long-term contracts (one or two years) and traditional payment methods tend to stay. Providing incentives for long-term commitments and reducing friction in digital billing could help reduce churn.
